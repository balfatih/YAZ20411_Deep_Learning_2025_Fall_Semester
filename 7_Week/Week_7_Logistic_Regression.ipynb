```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("/gdrive/My Drive/2025_2026_Derin_Ogrenme/Dry_Bean_Dataset.xlsx")
data
```

```python
data.drop_duplicates(inplace=True,keep='first')
```

```python
data['Class'].value_counts()
```

```python
x = data.drop(columns='Class')
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=20, test_size=0.2, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Decision Tree Model Değerlendirme

y_pred = lr_model.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

```python
from sklearn.metrics import matthews_corrcoef

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='weighted'))
print("Recall:", recall_score(y_test, y_pred, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
print("MCC: ", matthews_corrcoef(y_test, y_pred))
```

# Hyperparameter Optimization

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    # Düzenleme (Regularization) kuvvetinin tersi
    'C': np.logspace(-4, 4, 20),

    # Düzenleme Tipi
    'penalty': ['l1', 'l2', 'elasticnet', None],

    # Çözücü Algoritma
    # Not: 'l1' için 'liblinear' veya 'saga' kullanılmalı.
    # 'l2' için hepsi kullanılabilir ('lbfgs' varsayılan).
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],

    # Maksimum İterasyon Sayısı (ConvergenceWarning için artırıldı)
    'max_iter': [100, 500, 1000, 2000]
}
```

```python
random_search = RandomizedSearchCV(
    estimator=LogisticRegression(random_state=42), # random_state tekrarlanabilirlik sağlar
    param_distributions=param_grid,
    n_iter=50, # Rastgele 50 farklı kombinasyon dene
    cv=5,      # 5 katlı çapraz doğrulama kullan
    scoring='f1_weighted', # Veya problem tipinize uygun başka bir metrik
    verbose=1, # İşlem detaylarını göster
    n_jobs=-1, # Tüm çekirdekleri kullan (hızlandırmak için)
    random_state=42
)
```

```python
print("Randomized Search başlıyor...")
random_search.fit(X_train, y_train)

print("Randomized Search tamamlandı.")
```

```python
## En İyi Parametreler ve Skor

# En iyi parametreleri alma
best_params = random_search.best_params_
print(f"\n✨ En İyi Hiperparametreler:\n{best_params}")

# En iyi çapraz doğrulama skorunu alma
best_score = random_search.best_score_
print(f"En İyi CV Skoru ({random_search.scoring}): {best_score:.4f}")

## Optimize Edilmiş Model ile Tahmin ve Değerlendirme

# En iyi modeli al
best_lr_model = random_search.best_estimator_

# Test seti üzerinde tahmin yap
y_pred_optimized = best_lr_model.predict(X_test)

# Sınıflandırma Raporunu Görüntüle
print("\n--- Optimize Edilmiş Model Değerlendirme Raporu ---")
print(classification_report(y_test, y_pred_optimized, digits=4))
```
